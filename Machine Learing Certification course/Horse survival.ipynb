```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
animals=pd.read_csv("dataset/horse survival/horse.csv")
```

```python
target=animals["outcome"]
```

```python
target.unique()
```

```python
animals=animals.drop(["outcome"],axis=1)
```

```python
animals
```

```python
pd.get_dummies(animals["surgery"])
```

```python
# decision tree not work well with categorical variable

categorical_feature=[feature for feature in animals.columns if animals[feature].dtype=="O"]
```

```python
categorical_feature
```

```python
for category in categorical_feature:
    animals[category]=pd.get_dummies(animals[category])
```

```python
animals
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
X,y = animals.values,target.values
lb=LabelEncoder()
y=lb.fit_transform(y)
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)

```

```python
from sklearn.tree import DecisionTreeClassifier
print(x_train.shape)
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
```

```python
x_train=imp.fit_transform(x_train)
x_test=imp.fit_transform(x_test)
```

```python
classifier =DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
y_predict=classifier.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_predict,y_test)
```

```python
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
```

```python
classifier.fit(x_train,y_train)
y_predicted=classifier.predict(x_test)
accuracy=accuracy_score(y_predicted,y_test)
accuracy
```
